```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from CSV file
data = pd.read_csv("school_data.csv")

# Calculate total marks
data['total_marks'] = data[['Science', 'English', 'History', 'Maths']].sum(axis=1)

# Display the first few rows of the dataframe
print("First few rows of the data:")
print(data.head())

# Summary statistics
print("\nSummary statistics:")
print(data.describe())


# Histogram of marks for each subject
data[['Science', 'English', 'History', 'Maths']].hist(bins=10, figsize=(12, 8))
plt.suptitle('Histogram of Marks for Each Subject')
plt.show()

# Box plot of marks for each subject
data.boxplot(column=['Science', 'English', 'History', 'Maths'], figsize=(12, 8))
plt.title('Box Plot of Marks for Each Subject')
plt.ylabel('Marks')
plt.show()

# Gender-wise analysis
sns.boxplot(x='Gender', y='Science', data=data)
plt.title('Science Marks Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Science Marks')
plt.show()

sns.boxplot(x='Gender', y='English', data=data)
plt.title('English Marks Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('English Marks')
plt.show()

# Section-wise analysis
sns.barplot(x='Section', y='Maths', data=data)
plt.title('Average Maths Marks by Section')
plt.xlabel('Section')
plt.ylabel('Average Maths Marks')
plt.show()
```
